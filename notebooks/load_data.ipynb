```python
import pandas as pd
import mysql.connector
```

```python
df = pd.read_csv('../data/saas_funnel_raw.csv')
df.head()
```

```python
len(df)
```

```python

df = df.where(pd.notnull(df), None)
df = df.astype(object).where(pd.notnull(df), None)  # ← add this line
```

```python
df.head()
```

```python
# Connect to MySQL
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='1234',
    database='saas_funnel'
)
```

```python
cursor = conn.cursor()
print("Connected to MySQL successfully")

# Insert every row into MySQL
insert_query = """
INSERT INTO saas_funnel_raw (
    user_id, signup_date_raw, acquisition_channel, country_raw,
    device_raw, age, marketing_spend, session_count, visited_site,
    signed_up, completed_onboarding, onboarding_date_raw,
    activated_core_feature, activation_date_raw, started_trial,
    trial_start_date_raw, converted_to_paid, conversion_date_raw,
    plan_type, active_after_30_days
) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
"""


for index, row in df.iterrows():
    cursor.execute(insert_query, tuple(row))

conn.commit()
print("All rows inserted successfully")

# Verify count
cursor.execute("SELECT COUNT(*) FROM saas_funnel_raw")
count = cursor.fetchone()
print("Total rows in MySQL:", count[0])

cursor.close()
conn.close()
```
