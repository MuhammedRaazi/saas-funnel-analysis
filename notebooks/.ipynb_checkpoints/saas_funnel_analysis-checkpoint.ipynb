### Date Cleaning — Fixing Mixed Date Formats
This notebook connects to MySQL and fixes 5 date columns 
that have mixed formats using pandas pd.to_datetime()

```python
import pandas as pd           
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector 
from datetime import datetime
```

```python
conn = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password='1234',
    database = 'saas_funnel'
)
```

```python
df_dates = pd.read_sql("""
SELECT
    user_id,
    signup_date_raw,
    onboarding_date_raw,
    activation_date_raw,
    trial_start_date_raw,
    conversion_date_raw
FROM saas_funnel_cleaned
""", conn)
```

```python
df_dates['signup_date'] = pd.to_datetime(df_dates['signup_date_raw'],dayfirst=True,errors='coerce')
```

```python
df_dates['signup_date'].head(10)
```

```python
pd.__version__
```

```python
def parse_mixed_date(date_value):

    if pd.isna(date_value):
        return pd.NaT

    date_value = str(date_value).strip()

    formats = [
        '%d-%m-%Y',      # 26-05-2025
        '%m/%d/%Y',      # 06/22/2025
        '%Y-%m-%d',      # 2025-05-26
        '%Y/%m/%d',      # 2025/05/26
        '%b %d, %Y'      # Jul 04, 2025
    ]

    for fmt in formats:
        try:
            return datetime.strptime(date_value, fmt)
        except:
            pass

    return pd.NaT
```

```python
date_columns = [
    'signup_date_raw',
    'onboarding_date_raw',
    'activation_date_raw',
    'trial_start_date_raw',
    'conversion_date_raw'
]

for col in date_columns:
    new_col = col.replace('_raw', '')
    df_dates[new_col] = df_dates[col].apply(parse_mixed_date)
```

```python
df_dates[
    ['user_id',
     'signup_date',
     'onboarding_date',
     'activation_date',
     'trial_start_date',
     'conversion_date']
].head()
```

```python
df_analysis  = df_dates.copy()
```

```python
date_cols = [
    'signup_date',
    'onboarding_date',
    'activation_date',
    'trial_start_date',
    'conversion_date'
]

for col in date_cols:
    df_dates[col] = df_dates[col].where(
        pd.notna(df_dates[col]),
        None
    )
```

```python
cursor = conn.cursor()

update_query = """
UPDATE saas_funnel_cleaned
SET signup_date = %s,
    onboarding_date = %s,
    activation_date = %s,
    trial_start_date = %s,
    conversion_date = %s
WHERE user_id = %s
"""
```

```python
for _, row in df_dates.iterrows():

    cursor.execute(
        update_query,
        (
            None if pd.isna(row['signup_date']) else row['signup_date'].date(),
            None if pd.isna(row['onboarding_date']) else row['onboarding_date'].date(),
            None if pd.isna(row['activation_date']) else row['activation_date'].date(),
            None if pd.isna(row['trial_start_date']) else row['trial_start_date'].date(),
            None if pd.isna(row['conversion_date']) else row['conversion_date'].date(),
            int(row['user_id'])
        )
    )

conn.commit()

print("Dates updated successfully")
```

## Funnel Analysis

Analyze user progression through each stage of the SaaS funnel
to identify major drop-off points and conversion bottlenecks.

```python
# Calculating users at each stage
funnel_counts = {
    'Visited Site' : (df.visited_site == "Yes").sum(),
    'Signed Up' : (df['signed_up'] == "Yes").sum(),
    'Completed Onboarding' : (df['completed_onboarding']=="Yes").sum(),
    'Activated Core Feature' : (df['activated_core_feature']=="Yes").sum(),
    'Started Trial': (df['started_trial']=="Yes").sum(),
    'Converted to Paid': (df['converted_to_paid']=="Yes").sum()
}
```

```python
funnel_df = pd.DataFrame(funnel_counts.items(),columns = ['Stage','Users'])
```

```python
funnel_df
```

```python
# Calculate stage-to-stage conversion rates

funnel_df['Previous_Stage'] = funnel_df['Users'].shift(1)

funnel_df['Stage_Conversion_%'] = round(
    (funnel_df['Users'] / funnel_df['Previous_Stage']) * 100,
    2
)

funnel_df['Drop_Off_%'] = round(
    100 - funnel_df['Stage_Conversion_%'],
    2
)

funnel_df
```

```python
plt.figure(figsize=(12,6))

plt.bar(funnel_df['Stage'], funnel_df['Users'])

plt.title('SaaS Funnel Analysis')
plt.xlabel('Funnel Stage')
plt.ylabel('Users')

plt.xticks(rotation=20)

plt.show()
```

### Funnel Insights

Key Findings:

- Landing page performance is strong with 85.4% of visitors signing up.
- Significant drop-off occurs during onboarding and activation stages.
- The largest bottleneck is Trial to Paid conversion, where 60.4% of trial users do not become paying customers.
- Improving trial experience, pricing strategy, and product value communication and customer onboarding could         increase revenue.

## Retention Analysis

Analyze user retention after 30 days to understand long-term user engagement and product stickiness.

```python
# Count retained and non-retained users

retention_summary = (
    df['active_after_30_days']
    .value_counts()
)

retention_summary
```

```python
eligible_users = df['active_after_30_days'].isin(['Yes','No']).sum()

retained_users = (df['active_after_30_days'] == 'Yes').sum()

retention_rate = round(
    (retained_users / eligible_users) * 100,
    2
)

print("Retention Rate:", retention_rate, "%")
```

```python
retention_df = pd.DataFrame({
    'Status':['Retained','Churned'],
    'Users':[495,133]
})

retention_df
```

```python
plt.figure(figsize=(6,4))

plt.bar(
    retention_df['Status'],
    retention_df['Users']
)

plt.title('30-Day Customer Retention')
plt.xlabel('Status')
plt.ylabel('Users')

plt.show()
```

### Retention Insights

Key Findings:

- The 30-day retention rate is 78.82%, indicating strong customer engagement after conversion.
- Out of 628 paying customers, 495 remained active after 30 days.
- Only 133 customers became inactive within the first month, resulting in a relatively low churn rate.
- Strong retention suggests that users find value in the product after purchasing.
- Further improvements in onboarding, feature adoption, and customer success initiatives could increase long-term retention and customer lifetime value.

## Cohort Analysis

```python
cohort = pd.read_sql("""
SELECT DATE_FORMAT(signup_date,"%Y-%m") AS signup_month,
        COUNT(*) AS Totol_Users,
        SUM(CASE WHEN active_after_30_days = "Yes" then 1 else 0 end) AS retained_users,
        ROUND((SUM(CASE WHEN active_after_30_days = "Yes" then 1 else 0 end)/count(*))*100,2) AS retention_rate
        FROM saas_funnel_cleaned
        where signup_date is not null
        GROUP BY signup_month
        ORDER BY signup_month
""",conn)
```

```python
cohort
```

```python
cohort_pivot = cohort.set_index('signup_month')[['retention_rate']]
```

```python
cohort_pivot
```

```python
plt.figure(figsize=(4, 10))

sns.heatmap(cohort_pivot,annot=True,fmt='.1f',cmap='YlOrRd',linewidths=0.5,
    cbar_kws={'label': 'Retention Rate (%)'}
)

plt.title('Monthly Cohort Retention Rate')
plt.xlabel('Metric')
plt.ylabel('Signup Month')
plt.tight_layout()
plt.show()
```

##  Insights
* January, June, December 2024 and January 2025 had good retention — use those strategies in upcoming months
* Overall retention varies from 5% to 12% — not bad (Industry average for SaaS 30-day retention is 6-8%)
* After 2024, total users increased but retention rate decreased
* late 2025 cohorts show fewer users -- not a problem, just less time to convert since data ends Dec 2025

## Statistical Analysis

```python
contingency_table = pd.crosstab(df['device'],df['activated_core_feature'])
```

```python
contingency_table
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2, p_value, dof, expected = chi2_contingency(contingency_table)
print("Chi-square statistic:", round(chi2, 4))
print("P-value:", round(p_value, 4))
print("Degrees of freedom:", dof)
```

### Chi-Square Test Results

* A Chi-Square Test of Independence was conducted to evaluate whether device type influences core feature activation.
* The test produced a chi-square statistic of 0.5041 and a p-value of 0.918.
    Since the p-value is substantially greater than the significance level of 0.05, the null hypothesis could not be rejected.
    This indicates that there is no statistically significant relationship between device type and activation behavior.
* The small differences observed in activation counts across devices are likely due to random variation 
  rather than a true device-specific effect.

Business Recommendation:

Activation improvement efforts should focus on product experience, onboarding, and user engagement
rather than device-specific optimization.


#### Z-Test: Organic Search vs Paid Ads

Compare conversion rates between Organic Search and Paid Ads
to determine whether the observed difference is statistically significant.

```python
df.head()
```

```python
# Total users and converted users by acquisition channel
z_df = (
    df.groupby('acquisition_channel').agg(
          total_users=('user_id', 'count'),
          converted_users=('converted_to_paid',lambda x: (x == 'Yes').sum())
      )
      .reset_index()
)
```

```python
z_df = z_df[z_df.acquisition_channel.isin(["Organic Search","Paid Ads"])]
```

```python
z_df
```

```python
# Calculate conversion rate for each channel

z_df['conversion_rate'] = round(
    (z_df['converted_users'] / z_df['total_users']) * 100,
    2
)

z_df
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = [188, 152]

nobs = [1995, 1646]
```

```python
z_stat, p_value = proportions_ztest(
    count,
    nobs
)

print("Z-statistic:", round(z_stat, 4))
print("P-value:", round(p_value, 4))
```

### Z-Test Results: Organic Search vs Paid Ads

Key Findings:

- Organic Search conversion rate was 9.42%, while Paid Ads conversion rate was 9.23%.
- A two-proportion Z-test was performed to determine whether the observed difference was statistically significant.
- The test produced a p-value of 0.8453, which is greater than the significance level of 0.05.
- Therefore, the null hypothesis could not be rejected.
- The difference in conversion rates is likely due to random variation rather than a true performance difference.
- Based on this analysis, there is no evidence that Organic Search outperforms Paid Ads in terms of conversion rate.

## Pearson Correlation Analysis

Evaluate the relationship between marketing spend and subscription revenue to determine whether increased marketing investment
is associated with higher revenue generation.

```python
correlation = df['marketing_spend_cleaned'].corr(
    df['subscription_revenue']
)

print("Correlation:", round(correlation, 4))
```

### Pearson Correlation Results

Key Findings:

- Pearson correlation between marketing spend and subscription revenue was -0.0146.
- The correlation coefficient is very close to zero, indicating no meaningful linear relationship.
- Higher marketing spend does not consistently lead to higher subscription revenue at the individual user level.
- Revenue appears to be influenced by additional factors such as customer behavior, plan selection, and conversion outcomes.
- Marketing investment alone is not a strong predictor of subscription revenue in this dataset.

## Statistical Analysis Summary

Three statistical tests were conducted to validate key business assumptions and identify meaningful relationships within the SaaS funnel.

### Chi-Square Test: Device Type vs Feature Activation
- Chi-square statistic: 0.5041
- P-value: 0.9180
- Result: Failed to reject the null hypothesis.
- Conclusion: Device type and feature activation are independent. Activation behavior does not significantly differ across devices.

### Z-Test: Organic Search vs Paid Ads Conversion Rate
- Organic Search conversion rate: 9.42%
- Paid Ads conversion rate: 9.23%
- P-value: 0.8453
- Result: Failed to reject the null hypothesis.
- Conclusion: No statistically significant difference exists between Organic Search and Paid Ads conversion performance.

### Pearson Correlation: Marketing Spend vs Subscription Revenue
- Correlation coefficient: -0.0146
- Result: No meaningful linear relationship.
- Conclusion: Marketing spend alone is not a strong predictor of subscription revenue at the individual customer level.

### Overall Business Insights
- Device optimization is unlikely to improve activation rates.
- Organic Search and Paid Ads perform similarly in terms of conversion effectiveness.
- Revenue generation appears to be influenced by factors beyond marketing spend, such as customer quality, product adoption, and subscription plan selection.
- Future optimization efforts should focus on onboarding, activation, retention, and conversion rather than device-specific strategies.
